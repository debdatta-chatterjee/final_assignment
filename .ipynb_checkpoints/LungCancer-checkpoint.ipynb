```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

Data Collection and Analysis

```python
#Loading the data from csv file to Pandas Dataframe
file_path = '/content/dataset.csv'
lung_cancer_data = pd.read_csv(file_path)
```

```python
# Step 1: Data Overview
print("Dataset Overview:")
print(lung_cancer_data.head())

# Summary statistics
print("\nSummary Statistics:")
print(lung_cancer_data.describe(include='all'))

# Missing values
print("\nMissing Values:")
print(lung_cancer_data.isnull().sum())

# Distribution of target variable (LUNG_CANCER)
lung_cancer_distribution = lung_cancer_data['LUNG_CANCER'].value_counts()
print("\nLung Cancer Distribution:")
print(lung_cancer_distribution)
```

```python
# Step 2: Visualizing the Distribution of the Target Variable (Lung Cancer)
plt.figure(figsize=(8, 6))
sns.countplot(x='LUNG_CANCER', data=lung_cancer_data, palette='Set2')
plt.title('Distribution of Lung Cancer Cases')
plt.xlabel('Lung Cancer')
plt.ylabel('Count')
plt.show()
```

```python
# Step 3: Analyzing the Gender Distribution and its Relation to Lung Cancer
# Gender distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='GENDER', data=lung_cancer_data, palette='Set2')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Gender vs Lung Cancer
plt.figure(figsize=(8, 6))
sns.countplot(x='GENDER', hue='LUNG_CANCER', data=lung_cancer_data, palette='Set1')
plt.title('Gender vs Lung Cancer')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Step 4: Analyzing the Impact of Age on Lung Cancer
# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(lung_cancer_data['AGE'], bins=20, kde=True, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Age vs Lung Cancer
plt.figure(figsize=(10, 6))
sns.boxplot(x='LUNG_CANCER', y='AGE', data=lung_cancer_data, palette='Set2')
plt.title('Age vs Lung Cancer')
plt.xlabel('Lung Cancer')
plt.ylabel('Age')
plt.show()
```

```python
lung_cancer_data_encoded = pd.get_dummies(lung_cancer_data, drop_first=True)

plt.figure(figsize=(12, 10))
correlation_matrix = lung_cancer_data_encoded.corr()

sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Step 6: Analyzing the Relationship Between Key Risk Factors and Lung Cancer
# Smoking vs Lung Cancer
plt.figure(figsize=(8, 6))
sns.countplot(x='SMOKING', hue='LUNG_CANCER', data=lung_cancer_data, palette='Set3')
plt.title('Smoking vs Lung Cancer')
plt.xlabel('Smoking')
plt.ylabel('Count')
plt.show()
```

```python
# Chronic Disease vs Lung Cancer
plt.figure(figsize=(8, 6))
sns.countplot(x='CHRONIC_DISEASE', hue='LUNG_CANCER', data=lung_cancer_data, palette='Set3')
plt.title('Chronic Disease vs Lung Cancer')
plt.xlabel('Chronic Disease')
plt.ylabel('Count')
plt.show()
```

```python
# Coughing vs Lung Cancer
plt.figure(figsize=(8, 6))
sns.countplot(x='COUGHING', hue='LUNG_CANCER', data=lung_cancer_data, palette='Set3')
plt.title('Coughing vs Lung Cancer')
plt.xlabel('Coughing')
plt.ylabel('Count')
plt.show()
```
