```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Data collection and Analysis

```python
## Load the dataset
file_path = '/content/election2024.csv'
election_data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
print("Dataset Overview:")
print(election_data.head())

# Summary statistics
print("\nSummary Statistics:")
print(election_data.describe(include='all'))

# Missing values
print("\nMissing Values:")
print(election_data.isnull().sum())

# Distribution of preferred candidates
election_distribution = election_data['candidate'].value_counts()
print("\nCandidate Preference Distribution:")
print(election_distribution)
```

Visualizing Key Distributions

```python
import matplotlib.pyplot as plt
#Candidate Preference Distribution
plt.figure(figsize=(10, 6))
sns.countplot(y='candidate', data=election_data, palette='Set2', order=election_data['candidate'].value_counts().index)
plt.title('Preferred Candidate Distribution')
plt.xlabel('Count')
plt.ylabel('Candidate')
plt.show()
```

```python
#Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(election_data['age'], bins=20, kde=True, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Gender Distribution
plt.figure(figsize=(10, 6))
sns.histplot(election_data['age'], bins=20, kde=True, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Analyzing Voting Distributions

```python
#Voting intention by age
plt.figure(figsize=(10, 6))
sns.boxplot(x='will_vote', y='age', data=election_data, palette='Set2')
plt.title('Voting Intention by Age')
plt.xlabel('Voting Intention')
plt.ylabel('Age')
plt.show()
```

```python
#Voting intention by Federal District
plt.figure(figsize=(12, 8))
sns.countplot(y='federal_district', hue='will_vote', data=election_data, palette='Set3')
plt.title('Voting Intention by Federal District')
plt.xlabel('Count')
plt.ylabel('Federal District')
plt.show()
```

Analyzing Media Consumption Habits


```python
#Television Usage by Voting Intention
plt.figure(figsize=(10, 6))
sns.countplot(x='television_usage', hue='will_vote', data=election_data, palette='Set1')
plt.title('Television Usage by Voting Intention')
plt.xlabel('Television Usage')
plt.ylabel('Count')
plt.show()
```

```python
#Internet Usage by Voting Intention
plt.figure(figsize=(10, 6))
sns.countplot(x='internet_usage', hue='will_vote', data=election_data, palette='Set1')
plt.title('Internet Usage by Voting Intention')
plt.xlabel('Internet Usage')
plt.ylabel('Count')
plt.show()
```

Analyzing the Impact of Education and Income

```python
#Education Level Distribution
plt.figure(figsize=(10, 6))
sns.countplot(y='education', data=election_data, palette='Set2', order=election_data['education'].value_counts().index)
plt.title('Education Level Distribution')
plt.xlabel('Count')
plt.ylabel('Education Level')
plt.show()
```

```python
#Income Level Distribution
plt.figure(figsize=(10, 6))
sns.countplot(y='income', data=election_data, palette='Set2', order=election_data['income'].value_counts().index)
plt.title('Income Level Distribution')
plt.xlabel('Count')
plt.ylabel('Income Level')
plt.show()
```

Correlation Matrix


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/election2024.csv'
try:
    election_data = pd.read_csv(file_path)
except FileNotFoundError:
    print("The file was not found. Please check the file path.")

# Check for missing values
missing_values = election_data.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Optionally, fill or drop missing values before proceeding
# election_data.fillna(method='ffill', inplace=True)  # Example of filling missing values

# Convert categorical variables to numerical for correlation matrix
election_data_encoded = pd.get_dummies(election_data, drop_first=True)

# Check for any potential issues with too many columns
print("Number of columns after encoding:", election_data_encoded.shape[1])

# Compute the correlation matrix
plt.figure(figsize=(14, 12))
correlation_matrix = election_data_encoded.corr()

# Generate the heatmap
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
